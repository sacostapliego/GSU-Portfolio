### Problem 7

Below is a distribuƟon for the number of visits to denƟsts in one year. 

|X | 0 | 1 | 2 | 3 | 4 |
|---|---|---|---|---|---|
| P(X) | 0.1 | 0.3 | 0.4 | 0.15| 0.05 | 

Determine the expected value, the variance, and the standard deviaƟon of the distribuƟon. 

```r
x=c(0,1,2,3,4)
p=c(0.1,0.3,0.4,0.15,0.05)
mean = sum(x * p)
print(paste("Mean/Expected Value:", mean))

var = sum((x-mean)^2 * p)
print(paste("Variance:", var))

std = sqrt(var)
print(paste("Standard Deviation:", std))
```
